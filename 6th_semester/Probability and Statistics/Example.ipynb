```python
import sympy
sympy.init_printing()
x = sympy.IndexedBase('x', positive=True)
i,n,t = sympy.symbols('i n t', positive=True)
pprint((n*2)/sympy.summation(1/x[i] - x[i], (i,0,n)))
```

Below the result is shown:
$$\frac{2n}{\sum_{i=0}^{n}-x_i + \frac1{x_i}}$$
