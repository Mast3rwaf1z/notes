* Problem 5:
The time between phone calls in a grocery shop is given by an exponential
random variable $X ∼ Exp(λ)$, with an unknown rate of arrivals λ. The pdf of
this exponential variable is written as
$f (x, λ) = λ exp(−λx)$<br>

(a) Find the maximum likelihood estimator of λ<br>
(b) The shop attendant has been writing down the time between calls, and so
far he has noted the next 10 values, which are given in hours:
$1, 4, 6, 1, 3, 10, 2, 6, 9, 2$
What is the probability that there are no calls for more than 5 hours?

    a.
The pdf is written in an unintuitive way so:

$f(x, \lambda) = \lambda \cdot e^{-\lambda x}$

The likelyhood function is defined by:

$\mathcal{L}_n(\theta) = \prod_{i=0}^n f(X_i;\theta)$

where $\theta$ will be the maximum likelyhood estimator

i will be solving the equation using sympy in python, the equation should look as follows:

$\hat{\theta} = \frac{\mathscr{d}\ln{\mathcal{L}_n(\theta)}}{\mathscr{d}\theta}$

because sympy isn't a fan of products, i'll simplify the equation:

$$\mathcal{l}(\lambda) = \sum_{i=0}^n \ln{\lambda} - \lambda x$$

$$\mathcal{l}(\lambda) = n \cdot \ln{\lambda} \sum_{i=0}^n \lambda x$$

and i'll let sympy calculate the rest

```python
def MLE():
    from sympy import Eq, diff, Product, symbols, Symbol, exp, log, IndexedBase, init_printing, pprint, summation
    from sympy.solvers import solve
    init_printing()
    i, n, lmb = symbols('i n lmb')
    x = IndexedBase('x')
    log_likelihood = n * log(lmb) - lmb * summation(x[i], (i,0,n))
    print('log likelihood function:')
    pprint(log_likelihood)
    log_mle = Eq(diff(log_likelihood, lmb).doit(), 0)
    print('mle:')
    pprint(solve(log_mle.simplify(), lmb))
    return solve(log_mle.simplify(), lmb)
log_mle = MLE()
```

So the result is:

$$\frac{n}{\sum_{i=0}^n x_i}$$

An exam includes a question with four possible responses. The question is
designed to be very difficult, with none of the four responses being obviously
wrong, yet with only one correct answer. The exam is taken by 100 students.
The teacher wants to test whether more people answered the question correctly
than would be expected just due to chance (i.e., if every student picked the

(a) Set up the hypotheses for the test<br>
(b) Of the 100 students, 33 correctly answer the question. Find the p-value
and make a decision about the null hypothesis at a 5% significance level.

    a.

    b.

to find the p-value we employ the following equation:

$\text{p-value} = \sum_{k=33}^{100} \overset{100}{k} p_0^k(1-p_0)^{100-k} = 0.04459632521268153$

or in python:

```python
from math import factorial as fact
n = 100
k = 33
p0 = .25
# mathy method
print(f'p-value: {sum([(fact(n)/(fact(i) * fact(n - i))) * pow(p0, i) * pow(1 - p0, n - i) for i in range(k, n)])}')

# library methody thingy
from scipy import stats
print(1 - stats.binom.cdf(32,100,.25))
```

we can reject the null hypothesis as the p-value is less than our significance level

```python
def MLE():
    from sympy import Eq, diff, Product, symbols, Symbol, exp, log, IndexedBase, init_printing, pprint, summation
    from sympy.solvers import solve, nsolve
    init_printing()
    i, n, t = symbols('i n t')
    x = IndexedBase('x')
    log_likelihood = (n * log(t + 1)) + summation(log(x[i] ** t), (i,0,n))
    print('log likelihood function:')
    pprint(log_likelihood)
    log_mle = Eq(diff(log_likelihood, t).doit(), 0)
    pprint(log_mle)
    print('mle:')
    pprint(solve(log_mle, t))
MLE()
```

This exercise can also be solved by hand:

likelihood:

$$\mathcal{L}(\theta) = \prod_{i=0}^{n} (\theta + 1) \cdot x^\theta$$

log likelihood:

$$\ln{\mathcal{L}(\theta)} = n * \ln{\theta + 1} + \sum_{i=0}^{n} \ln{x^\theta}$$

$$\ln{\mathcal{L}(\theta)} = n * \ln{\theta + 1} + \sum_{i=0}^{n} \theta * \ln{x}$$

$$\text{diff}(\ln{\mathcal{L}(\theta)}) = \frac{n}{\theta + 1} + \sum_{i=0}^{n} \ln{x}$$

$$\frac{n}{\theta + 1} + \sum_{i=0}^{n} \ln{x} = 0$$

$$-\frac{n}{\theta + 1} = \sum_{i=0}^{n} \ln{x}$$

$$-n = \sum_{i=0}^{n} \ln{x} * (\theta + 1)$$

$$\frac{-n}{\sum_{i=0}^{n} \ln{x}} = \theta + 1$$

$$\theta = \frac{-n}{\sum_{i=0}^{n} \ln{x}} - 1$$
